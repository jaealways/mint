```python
import sys
sys.path.append("..")

from fng.score import Score, FearGreed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_mcpi_volume = pd.read_pickle("../storage/df_raw_data/df_song_volume_17-01-01_23-12-31.pkl")
df_mcpi_index = pd.read_pickle("../storage/df_raw_data/df_price_17-01-01_23-12-31.pkl")
x, y = df_mcpi_index.to_numpy(), df_mcpi_volume.to_numpy()
```

```python
score = Score(x, y)
score_fng = FearGreed(x, y, score).compute()
```

```python
#뮤직카우 번호 num_mc에 입력
num_mc = 1016
date = df_mcpi_index.columns[-len(score_fng):]
index_list = df_mcpi_volume.index.tolist()
num = index_list.index(num_mc)
```

```python
# 위에서부터 가격, 거래량, 공포-탐욕지수
fig, axs = plt.subplots(3)
axs[0].plot(x[num, -score_fng.shape[1]:])
axs[1].plot(y[num, -score_fng.shape[1]:])
axs[2].plot(score_fng[num, :])
```
